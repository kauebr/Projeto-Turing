```python
def mms(vezes=10, intervalo=1):
    """
    Move o mouse aleatoriamente pela tela.

    Parâmetros:
    - vezes: número de movimentos aleatórios
    - intervalo: tempo (em segundos) entre cada movimento
    """
    largura_tela, altura_tela = pyautogui.size()

    for _ in range(vezes):
        x = random.randint(0, largura_tela - 1)
        y = random.randint(0, altura_tela - 1)
        duracao = random.uniform(0.1, 0.5)  # duração do movimento
        pp.moveTo(x, y, duration=duracao)
        time.sleep(intervalo)
```

```python

# Caminho do arquivo CNAE
cnae_path = r"C:\Users\kaue.silva\OneDrive - K8 Soluções em Pagamentos ltda\Escriba_C\Cordon Bleu\Base de Dados\Dezipados\F.K03200$Z.D41214.CNAECSV"

# Lendo o arquivo CNAE (sem cabeçalho e separador ";")
cnae_df = pd.read_csv(cnae_path, sep=';', encoding= 'latin1', header=None, names=['CODIGO CNAE', 'DESCRICAO CNAE'])

# Criando o dicionário de mapeamento CNAE
cnae_dict = cnae_df.set_index('CODIGO CNAE')['DESCRICAO CNAE'].to_dict()
```

```python
# Função para mapear o código CNAE principal
estabelecimentos['CNAE FISCAL PRINCIPAL'] = estabelecimentos['CNAE FISCAL PRINCIPAL'].map(cnae_dict)

# Função para mapear os códigos CNAE secundários (separados por vírgulas)
def mapear_cnaes(cnaes):
    if pd.isna(cnaes):  # Caso a célula seja NaN
        return cnaes
    return ', '.join([cnae_dict.get(int(cnae.strip()), cnae.strip()) for cnae in cnaes.split(',')])

estabelecimentos['CNAE FISCAL SECUNDÁRIA'] = estabelecimentos['CNAE FISCAL SECUNDÁRIA'].apply(mapear_cnaes)
```

```python

# Passo 1: Converter todos os e-mails para minúsculas, tratando valores nulos
estabelecimentos['CORREIO ELETRÔNICO'] = estabelecimentos['CORREIO ELETRÔNICO'].fillna('').str.lower()

# Adicionando a coluna 'Contador' ao DataFrame
estabelecimentos['Contador'] = (
    (
        estabelecimentos['CORREIO ELETRÔNICO'].str.contains('contab', na=False) |
        estabelecimentos['CORREIO ELETRÔNICO'].str.contains('contad', na=False)\
    ) 
)

# Exibir o DataFrame com a nova coluna
print(estabelecimentos[['CORREIO ELETRÔNICO', 'Contador']])
```

```python
# Passo 1: Converter a coluna 'DATA DE INÍCIO ATIVIDADE' para o formato datetime
estabelecimentos['DATA DE INÍCIO ATIVIDADE'] = pd.to_datetime(estabelecimentos['DATA DE INÍCIO ATIVIDADE'], format='%Y%m%d')

# Passo 2: Calcular a diferença em meses entre a data atual e a 'DATA DE INÍCIO ATIVIDADE'
# Obtendo a data atual
data_atual = pd.to_datetime('today')

# Calculando a diferença em meses
estabelecimentos['Tempo de Atividade (meses)'] = ((data_atual.year - estabelecimentos['DATA DE INÍCIO ATIVIDADE'].dt.year) * 12) + (data_atual.month - estabelecimentos['DATA DE INÍCIO ATIVIDADE'].dt.month)

# Exibir o DataFrame com a nova coluna
print(estabelecimentos[['DATA DE INÍCIO ATIVIDADE', 'Tempo de Atividade (meses)']])
```
