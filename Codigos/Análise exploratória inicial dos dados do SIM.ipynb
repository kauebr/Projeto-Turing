```python
# Total de entradas
total = len(df)

# Verifica√ß√µes
nulos = df['MUNIRES'].isnull().sum()
vazios = (df['MUNIRES'] == '').sum()
nan_texto = (df['MUNIRES'].astype(str).str.upper() == 'NAN').sum()

# Num√©ricos (exclui NaN e vazios antes de checar)
numericos = pd.to_numeric(df['MUNIRES'], errors='coerce').notna().sum()

# Percentuais
print("An√°lise da coluna MUNIRES (%):")
print(f"- Valores nulos (NaN/None): {nulos / total:.2%}")
print(f"- Valores em branco (''): {vazios / total:.2%}")
print(f"- Texto 'NaN': {nan_texto / total:.2%}")
print(f"- Valores num√©ricos v√°lidos: {numericos / total:.2%}")
```
