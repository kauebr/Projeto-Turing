```python
# Importando as bibliotecas necessárias
import pandas as pd
```

```python
# Lendo o arquivo CSV em um DataFrame
df = pd.read_csv(r"C:\... POR ENDEREÇO AQUI", sep=';')
```

```python
# Removendo Coluna 'contador'
df = df.drop(columns=['contador'])
```

```python
# Parsing da coluna 'TIPOBITO' para o valor 'Fatalidade' = 2
df['TIPOBITO'] = 'Fatalidade'
```

```python
# Parsing da coluna 'SEXO' para valores descritivos
df['SEXO'] = df['SEXO'].replace({
    1: 'Masculino',
    2: 'Feminino',
    0: 'Ignorado'
})
```

```python
df.to_csv(r"C:\... POR ENDEREÇO AQUI", sep=';', index=False)
```
